```python
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from wordcloud import WordCloud
import matplotlib.pyplot as plt

file_path = 'combined.txt'
with open(file_path, 'r') as file:
    text = file.read()


tokenized_text = word_tokenize(text.lower())


tf_word_freq = FreqDist(tokenized_text)


documents = text.split('\n')
df_word_freq = FreqDist()
for doc in documents:
    words = set(word_tokenize(doc.lower()))
    df_word_freq.update(words)


#This is for Term Frequency
tf_wordcloud = WordCloud(width=800, height=400, background_color='white')
tf_wordcloud.generate_from_frequencies(tf_word_freq)

#THis is for Document Frequency
#Word Frequency
df_wordcloud = WordCloud(width=800, height=400, background_color='white')
df_wordcloud.generate_from_frequencies(df_word_freq)

```

```python
plt.figure(figsize=(12, 6))

plt.imshow(tf_wordcloud, interpolation='bilinear')
plt.title('Term Frequency Word Cloud')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.imshow(df_wordcloud, interpolation='bilinear')
plt.title('Document Frequency Word Cloud')
plt.axis('off')
plt.tight_layout()
plt.show()
```
